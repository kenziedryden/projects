```python
cities = [
    "New York", "Los Angeles", "Chicago", "Toronto", "Vancouver",
    "Mexico City", "Buenos Aires", "São Paulo", "Rio de Janeiro", "Santiago",
    "Lima", "Bogotá", "Caracas", "Quito", "Panama City",
    "Havana", "San Jose", "Guatemala City", "Kingston", "Port-au-Prince",
    "London", "Paris", "Berlin", "Rome", "Madrid",
    "Vienna", "Amsterdam", "Brussels", "Dublin", "Lisbon",
    "Warsaw", "Budapest", "Prague", "Athens", "Stockholm",
    "Copenhagen", "Oslo", "Helsinki", "Zurich", "Geneva",
    "Luxembourg", "Ljubljana", "Zagreb", "Sofia", "Belgrade",
    "Bucharest", "Tallinn", "Riga", "Vilnius", "Bratislava",
    "Moscow", "Saint Petersburg", "Istanbul", "Kyiv", "Minsk",
    "Reykjavik", "Sarajevo", "Skopje", "Podgorica", "Tirana",
    "Dubai", "Abu Dhabi", "Riyadh", "Jeddah", "Tel Aviv",
    "Beirut", "Amman", "Cairo", "Casablanca", "Algiers",
    "Tunis", "Tripoli", "Nairobi", "Accra", "Lagos",
    "Addis Ababa", "Dakar", "Cape Town", "Johannesburg", "Kinshasa",
    "Tokyo", "Osaka", "Seoul", "Beijing", "Shanghai",
    "Hong Kong", "Taipei", "Bangkok", "Kuala Lumpur", "Singapore",
    "Jakarta", "Manila", "Hanoi", "Ho Chi Minh City", "Phnom Penh",
    "Sydney", "Melbourne", "Auckland", "Wellington", "Suva",
    "Islamabad", "Delhi", "Mumbai", "Dhaka", "Kathmandu"
]

# Dictionary to store the coordinates
coordinates = {}

# Load the worldcities.csv file with pandas
df = pd.read_csv('worldcities.csv')

# List to collect cities that are not found in the dataset
missing_cities = []

# Populate coordinates for each city in the list
for city in cities:
    # Locate the row in the DataFrame for the specific city
    city_data = df[df['city'] == city]
    if not city_data.empty:
        # Get the latitude and longitude and add to the dictionary
        coordinates[city] = (city_data.iloc[0]['lat'], city_data.iloc[0]['lng'])
        #print(city, coordinates[city])
    else:
        print(city, 'not in data set')
        missing_cities.append(city)

# Remove missing cities from the original list
cities = [city for city in cities if city not in missing_cities]
```

```python
# Calculate distances between each pair of cities
n = len(cities)
distances = np.zeros((n, n))
for i in range(n):
    for j in range(i + 1, n):
        city1, city2 = cities[i], cities[j]
        dist = geodesic(coordinates[city1], coordinates[city2]).kilometers
        distances[i][j] = distances[j][i] = dist
```
